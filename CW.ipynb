```python
import random
import numpy as np
class OnePointCrossover:
    def __init__(self, number_of_populations, ind_per_pop, gene_length):
        self.number_of_populations = number_of_populations
        self.ind_per_pop = ind_per_pop
        self.gene_length = gene_length
        self.R = self.randomize_r(gene_length)
        self.whole_population = self.initate_individuals_in_populations( number_of_populations, ind_per_pop,gene_length)
        self.fittest_fitness = self.R[gene_length,gene_length] * (pow(2,gene_length) + pow(2,gene_length))
        self.mutation_rate = 1/(2*gene_length)
        
    def randomize_r(self, gene_length):
        assigned_values = {}
        uniform_array= np.random.uniform(0.51,1,(gene_length+1)*(gene_length+1))
        uniform_array.sort()
        uniform_array[-1] = 1.00
        index = 0
        for i in list(range(0,gene_length+1)):
                for j in list(range(0,gene_length+1)):
                    assigned_values[i,j] = uniform_array[index]
                    index = index + 1
        return assigned_values
    
    def initate_individuals_in_populations(self,number_of_populations, individuals_per_population,gene_length):
        populations = {}
        for i in list(range(number_of_populations)):
            population=[]
            for j in list(range(number_of_populations)):
                population.append(self.generate_random_ind(gene_length))
            populations[f"population{i}"] = population
        return populations

    
    def generate_random_ind(self, gene_length):
        random_ind = []
        binary = ['0','1']
        while len(random_ind) != 2* gene_length:
            random_ind.append(random.choice(binary))
        return ''.join(random_ind)

    def fitness_function(self, genotype):
        left_gene, right_gene = self.genotype_to_genes(genotype)
        i = left_gene.count('1')
        j = right_gene.count('1')
        return self.R[i,j] * (pow(2,i) + pow(2,j))
    
    def genotype_to_genes(self, genotype):
        left_gene = genotype[0:int(len(genotype)/2)]
        right_gene = genotype[int(len(genotype)/2):int(len(genotype))]
        return left_gene, right_gene

    
    def one_point_crossover(self, parent1, parent2):
        crossover_point = random.choice(list(range(1,len(parent1))))
        child=[]
        print(crossover_point)
        for el in list(range(0,crossover_point)):
#             print(el)
#             print(parent1[0])
            child.append(parent1[el])
        for el in list(range(crossover_point,len(parent2))):
            child.append(parent2[el])
        return child


    def mutate(self, population):
        binary = ['0','1']
        for individual in population:
            for el in individual:
                if(random.uniform(0, 1)< self.mutation_rate):
                    el = random.choice(binary)
                    print("Mutation occurred")
        return population
                
    def migrate(self, whole_population, ind_per_pop):
        migrants = []
        for el in whole_population:
            ind_pos = list(range(0,ind_per_pop))
            random_individual = random.choice(ind_pos)
#             print(random_individual)
            migrants.append(whole_population[el][random_individual])
            whole_population[el].remove(whole_population[el][random_individual])
#         print("M")
        print(migrants)
#         print(whole_population)
        index = 0
        available_populations = list(range(0,ind_per_pop))
        for el in migrants:
            print(el)
            print(set(available_populations))
            print(set(([migrants.index(el)])))
            if(set(available_populations) == set(([migrants.index(el)]))):
                whole_population[f'population{migrants.index(el)}'].append(el)
            else:
                temp_av=list(set(available_populations)-set(([migrants.index(el)])))
                print(temp_av)
        #         available_populations.remove(index)
    #             print(available_populations)
                random_new_population = random.choice(temp_av)
                available_populations.remove(random_new_population)
                whole_population[f'population{random_new_population}'].append(el)
            index=index+1
        return whole_population

    def check_for_fittest(self, whole_population, generation):
        print(whole_population)
        for pop in whole_population:
            for geno in whole_population[pop]:
                if self.fitness_function(geno) == self.fittest_fitness:
                    print(f'FITTEST FITNESS FOUNDED IN GENERATION {generation}')
                    print(geno)
                    return True
        return False
    
    def proportionate_fitness_selection(self, population):
        fitness_sum = 0
        for el in population:
            fitness_sum = fitness_sum + self.fitness_function(el)

        found = False
        while(found is False):
            random_parent = random.choice(population)
            if(random.uniform(0, 1) < self.fitness_function(random_parent)/fitness_sum):
                parent1 = random_parent
                found=True
        found = False
        while(found is False):
            random_parent = random.choice(population)
            if(random.uniform(0, 1) < self.fitness_function(random_parent)/fitness_sum):
                parent2 = random_parent
                found=True

        return parent1,parent2

    def find_fittest_in_population(self, population):
        fittest = population[0]
        for el in population[1:]:
            if(self.fitness_function(el)>self.fitness_function(fittest)):
                fittest = el
        return fittest
            
            
            

    def simulation(self):
        generation = 0
        while self.check_for_fittest(self.whole_population, generation) is not True:
            print(f'Generation {generation} started')
#             Migrate populations
            self.migrate(self.whole_population, self.ind_per_pop)
            
            for pop in self.whole_population:
#                 print(pop)
                new_population = []
#                 Keep fittest individual
                new_population.append(self.find_fittest_in_population(self.whole_population[pop]))
                                      
                for el in range(len(self.whole_population[pop])-1):
#                     Proportionate selection
                    parent1, parent2 = self.proportionate_fitness_selection(self.whole_population[pop])
#                     One point crossover to generate child
                    child = self.one_point_crossover(parent1, parent2)
                    new_population.append(child)
#                 Mutating new population
                self.whole_population[pop] = self.mutate(new_population)
            generation = generation + 1
            print(f'Generation {generation} ended without success')

                    
                    
            
            
            
            
    
    
    
        
        
    
```
